```python
# Initial imports
import os
import pandas as pd
from pathlib import Path
from newsapi import NewsApiClient
from dotenv import load_dotenv
load_dotenv()

# Retrieve the News API key
api_key = os.getenv("NEWS_API_KEY")

# Create the newsapi client
newsapi = NewsApiClient(api_key=api_key)

# Verification of API key for NewsApi
# Should be 'str' and 32

print(type(api_key))
print(len(api_key))
```

```python
newsapi.get_everything(q="Democrat",language="en")
```

```python
def create_news_df(news):
    """
    create_df takes in a json file created by NewsAPI and returns
    a dataframe of key information on the queried subject.
    """
    
    #Creating an open list of dictionaries to create a Dataframe later on.
    articles = []
    
    #Going through the articles found and using a Try/Except block to not break code.
    for article in news:
        
        try:
            
            title = article["title"]
            description = article["description"]
            text = article["content"]
            date = article["publishedAt"][:10]

            articles.append({
                "Title": title,
                "Description": description,
                "Text": text,
                "Date": date,
            })
            
        except AttributeError:
            
            pass

    return pd.DataFrame(articles)
```

```python
bitcoin_news_df = create_news_df(newsapi.get_everything(q="bitcoin",language="en")['articles'])
```
